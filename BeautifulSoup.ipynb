```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random


query = input("search : ")
url = 'https://shopee.tw/search?keyword=' + query
header = {'User-Agent':'Googlebot'}

r = requests.get(url, headers=header)
print(r.status_code)
print(r.text)
soup = BeautifulSoup(r.text, 'html.parser')
print(soup.prettify)
```
